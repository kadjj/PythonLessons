```python
#replacing the problematic/missing symbols

austin_weather['VisibilityLowMiles'] = austin_weather['VisibilityLowMiles'].replace('-', np.nan).astype('float')
austin_weather['VisibilityHighMiles'] = austin_weather['VisibilityHighMiles'].replace('-', np.nan).astype('float')
austin_weather['VisibilityAvgMiles'] = austin_weather['VisibilityAvgMiles'].replace('-', np.nan).astype('float')
```

```python
n = 50

# plot initiation
fig, ax = plt.subplots(2, 1, sharex = True)  # means 2 rows, 1 column, sharex means that the parameters share x axis

# create the line charts (plt.plot) for TempHighF, TempAvgF, TempLowF

ax[0].plot(austin_weather.index[0:n], austin_weather['TempHighF'][0:n], linestyle = '--')
ax[0].plot(austin_weather.index[0:n], austin_weather['TempAvgF'][0:n])
ax[0].plot(austin_weather.index[0:n], austin_weather['TempLowF'][0:n], linestyle = '--')

ax[1].plot(austin_weather.index[0:n], austin_weather['VisibilityHighMiles'][0:n], linestyle = '--')
ax[1].plot(austin_weather.index[0:n], austin_weather['VisibilityAvgMiles'][0:n])
ax[1].plot(austin_weather.index[0:n], austin_weather['VisibilityLowMiles'][0:n], linestyle = '--')


ax[0].set_title('Temperatures in Austin')

ax[1].set_xlabel('Date')
ax[0].set_ylabel('Termperature(F)')
ax[1].set_ylabel('Visibility (Miles)')


ax[1].xaxis.set_tick_params(rotation = 45)


# display the plot
plt.show()
```

```python
population = pd.read_csv('population.csv')
```

```python
population.columns
```

```python
population.head()
```

```python
population['Sex'].value_counts(dropna = False)
```

```python
grouped = population.groupby('Sex')
both, female, male = [grouped.get_group(x) for x in grouped.groups]
```

```python
fix, ax = plt.subplots()

ax.bar(both['City'], both['Value'])
ax.xaxis.set_tick_params(rotation = 90)
ax.set_title('Population of Poland')
ax.set_ylabel('Population (mln)')

plt.show()
```

```python
fix, ax = plt.subplots()

ax.bar(female['City'], female['Value'], label = 'female', bottom = male['Value'],color = 'purple')
ax.bar(male['City'], male['Value'], label = 'male')

ax.xaxis.set_tick_params(rotation = 90)
ax.set_title('Population of Poland')
ax.set_ylabel('Population (mln)')

ax.legend()

plt.show()
```

```python
# scatter plot

fig, ax = plt.subplots()

ax.scatter(austin_weather['TempAvgF'], austin_weather['HumidityAvgPercent'])
ax.set_xlabel('Temperature (F)')
ax.set_ylabel('Humidity')
                          

plt.show()
```

```python


austin_weather['HumidityLowPercent'] = austin_weather['HumidityLowPercent'].replace('-', np.nan).astype('float')
austin_weather['HumidityAvgPercent'] = austin_weather['HumidityAvgPercent'].replace('-', np.nan).astype('float')
austin_weather['HumidityHighPercent'] = austin_weather['HumidityHighPercent'].replace('-', np.nan).astype('float')
```

```python
# scatter plot

fig, ax = plt.subplots()

ax.scatter(austin_weather['TempAvgF'], austin_weather['HumidityAvgPercent'])


ax.set_xlabel('Temperature (F)')
ax.set_ylabel('Humidity')
                          

plt.show()
```

```python
austin_weather['Visibility_grouped'] = pd.cut(austin_weather['VisibilityAvgMiles'], bins = 3)
```

```python
austin_weather['Visibility_grouped']
```

```python
grouped = austin_weather.groupby('Visibility_grouped')
low, middle, high = [grouped.get_group(x) for x in grouped.groups]
 
```

```python
# scatter plot

fig, ax = plt.subplots()

ax.scatter(low['TempAvgF'], low['HumidityAvgPercent'], label = 'Low')
ax.scatter(middle['TempAvgF'], middle['HumidityAvgPercent'], label = 'Middle')
ax.scatter(high['TempAvgF'], high['HumidityAvgPercent'], label = 'High')



ax.set_xlabel('Temperature (F)')
ax.set_ylabel('Humidity')
ax.legend()
                          

plt.show()
```
